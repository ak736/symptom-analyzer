## **INSERT THE DATA SET**

```python
df=pd.read_csv('/content/sample_data/kidney_disease.csv')
```

```python
df.drop("id", axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'classs']
```

```python
df.head()
```

```python
df.info()
```

```python
df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'], errors='coerce')
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'], errors='coerce')
df['red_blood_cell_count'] = pd.to_numeric(df['red_blood_cell_count'], errors='coerce')
```

```python
df.info()
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
num_cols = [col for col in df.columns if df[col].dtype != 'object']
```

```python
df['diabetes_mellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)

df['coronary_artery_disease'] = df['coronary_artery_disease'].replace(to_replace = '\tno', value='no')

df['classs'] = df['classs'].replace(to_replace = {'ckd\t': 'ckd', 'notckd': 'not ckd'})
```

```python
df['classs'] = df['classs'].map({'ckd': 1, 'not ckd': 0})
df['classs'] = pd.to_numeric(df['classs'], errors='coerce')
```

```python
cols = ['diabetes_mellitus', 'coronary_artery_disease', 'classs']

for col in cols:
    print(f"{col} has {df[col].unique()} values\n")
```

```python
missing=df.isna().sum().sort_values(ascending=False)
print(missing)
```

```python
df.dropna().shape
```

```python
numerical=[]
for col in df.columns:
    if df[col].dtype=="float64":
        numerical.append(col)
print(numerical)
```

```python
categorical=[]
for col in df.columns:
    if df[col].dtype=="object":
        categorical.append(col)
print(categorical)
```

```python
for col in df.columns:
    if col in numerical:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
for col in categorical:
    df[col]=encoder.fit_transform(df[col])
```

# **SPLIT THE DATA INTO TRAIN AND TEST**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df[['age', 'blood_pressure', 'specific_gravity','red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia']],df.classs,test_size=0.2)
```

## **SVM**

```python
from sklearn.svm import SVC
```

```python
model=SVC(kernel='linear',probability=True)
```

```python
model.fit(X_train,y_train)
```

```python
svc_acc_score=model.score(X_test,y_test)
```

```python
svc_acc_score
```

```python
y_pred_sk=model.predict(X_test)
print(classification_report(y_test,y_pred_sk))
```
